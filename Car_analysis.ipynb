```python
def car_overview_dict_creator(overview):
    overview_dict = json.loads(overview.replace("'", '"').replace('None', "1"))
    all_overviews = {}
    for dictionary in overview_dict['top']:
        all_overviews.update({dictionary['key']: dictionary['value']})
    return all_overviews
```

```python
def car_feature_dict_creator(feature):
  all_features = {}
  feature_dict = json.loads(feature.replace("'", '"').replace('None', "1"))
  for dictionary in feature_dict['top']:
    all_features.update({dictionary['value']: True})
    for section in feature_dict.get('data', []):
      for item in section.get('list', []):
        all_features.update({item['value']: True})
  return all_features
```

```python
def car_spec_dict_creator(spec):
    all_specs = {}
    specs_dict = json.loads(spec.replace("'", '"').replace('None', "null"))
    for dictionary in specs_dict.get('top', []):
      all_specs.update({dictionary['key']: dictionary['value']})
    # Extract nested specifications
    for dictionary in specs_dict.get('data', []):
      for item in dictionary.get('list', []):
        all_specs.update({item['key'] : item['value']})
    return all_specs
```

```python
def null_value_imputer(df):
    # Handling null values
    for i in df.columns:
        if (df[i].isnull().sum() / len(df)) * 100 >=50:
            df.drop(i, axis=1, inplace=True)
        elif df[i].dtypes in ["int64", "float64"]:
            if df[i].skew() < 0.4 and df[i].skew() > -0.4:
                df[i].fillna(df[i].mean(), inplace=True)
            else:
                df[i].fillna(df[i].median(), inplace=True)
        else:
            df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
def convert_to_lakhs(value):
    """
    Converts a price string (e.g., ₹2.10Crore, ₹8 Lakh) into a numeric value in lakhs.
    """
    value = value.lower().replace("₹", "").replace(",", "").strip()  # Remove ₹ and commas
    
    if "crore" in value:
        # Convert Crore to Lakhs
        return float(value.replace("crore", "").strip()) * 100
    elif "lakh" in value:
        # Convert Lakh directly
        return float(value.replace("lakh", "").strip())
    elif value.isnumeric():
        # Assume plain numbers are in rupees and convert to lakhs
        return float(value) / 100000
    else:
        return None  # Handle invalid cases
```

```python
def clean_maxpower(value):
    # Remove the rpm part and anything that follows '@' or 'at'
    value = value.split('@')[0].split(' at')[0].lower().strip()
    
    # Extract the numeric value and handle both bhp and PS
    if 'bhp' in value:
        power = value.replace('bhp', '').strip()
    elif 'ps' in value:
        power = value.replace('ps', '').strip()
        power = float(power) * 0.9863  # Convert PS to bhp
    else:
        power = None  # If the value doesn't match expected format
    
    # Return converted power value as float, if None return NaN
    try:
        return float(power) if power is not None else np.nan  # Use np.nan for missing values
    except ValueError:
        return np.nan  # Handle unexpected formats
```

```python
# Function to categorize tyre types
def categorize_tyre_type(value):
    # Lowercase, remove extra spaces and split by common delimiters
    value = value.lower().replace('.', ',').replace('-', '').strip()
    components = [part.strip() for part in value.split(',')]

    # Standardize terms
    standardized = []
    for component in components:
        if 'tubeless' in component:
            standardized.append('Tubeless')
        elif 'radial' in component:
            standardized.append('Radial')
        elif 'runflat' in component or 'run-flat' in component:
            standardized.append('Runflat')
        else:
            standardized.append('Other')  # Catch unexpected values

    # Combine and deduplicate categories
    return ', '.join(sorted(set(standardized)))
```

```python
obj_df['Tyre Type'] = obj_df['Tyre Type'].apply(categorize_tyre_type)
```

```python
# Function to clean and categorize drive types
def categorize_drive_type(value):
    # Lowercase and strip extra spaces
    value = value.lower().strip()

    # Standardize terms using mapping
    if 'fwd' in value or 'front wheel' in value:
        return 'FWD'  # Front Wheel Drive
    elif 'rwd' in value or 'rear wheel' in value:
        return 'RWD'  # Rear Wheel Drive
    elif 'awd' in value or 'all wheel' in value:
        return 'AWD'  # All Wheel Drive
    elif '4wd' in value or '4x4' in value:
        return '4WD'  # Four Wheel Drive
    elif '2wd' in value or '4x2' in value or 'two wheel' in value:
        return '2WD'  # Two Wheel Drive
    else:
        return 'Other'  # Catch unexpected values
obj_df['Drive Type'] = obj_df['Drive Type'].apply(categorize_drive_type)
```

```python
# Function to categorize fuel supply systems
def categorize_fuel_supply(value):
    value = value.lower().strip()  # Normalize case and strip spaces

    # Mapping common terms to categories
    if 'mpfi' in value or 'multi point' in value or 'multipoint' in value:
        return 'MPFI'
    elif 'crdi' in value or 'common rail' in value:
        return 'CRDI'
    elif 'direct injection' in value or 'di' in value:
        return 'Direct Injection'
    elif 'efi' in value or 'electronic fuel injection' in value:
        return 'EFI'
    elif 'gdi' in value or 'gasoline direct injection' in value:
        return 'GDI'
    elif 'cng' in value:
        return 'CNG'
    elif 'electric' in value:
        return 'Electric'
    else:
        return 'Other'  # Catch-all for uncommon systems
# Apply the categorization function
obj_df['Fuel Suppy System'] = obj_df['Fuel Suppy System'].apply(categorize_fuel_supply)
```

```python
def categorize_gearbox(value):
    # Normalize the value to lowercase for easier matching
    value = str(value).lower()
    
    # Manual Transmission Categories
    if 'manual' in value or 'speed' in value and 'dct' not in value and 'cv' not in value:
        return 'Manual Transmission'
    
    # Automatic Transmission Categories
    elif 'automatic' in value or 'cv' in value or 'iv' in value or 'fully' in value:
        return 'Automatic Transmission'
    
    # Semi-Automatic / Dual Clutch Transmission (DCT) Categories
    elif 'dct' in value or 'steptronic' in value or 'ds' in value:
        return 'Dual-Clutch Transmission (DCT)'
    
    # Other categories
    else:
        return 'Other Transmission'
obj_df['Gear Box'] = obj_df['Gear Box'].apply(categorize_gearbox)
```

```python
import re

def convert_torque(value):
    if pd.isna(value):  # Handle NaN
        return None
    value = value.lower()  # Convert to lowercase for uniformity
    # Extract the numeric part and unit
    match = re.search(r'([\d.]+)\s*(nm|kgm)', value)
    if match:
        num, unit = match.groups()
        num = float(num)
        if unit == 'kgm':  # Convert kgm to Nm
            num *= 9.8
        return num
    return None  # Return None for invalid entries
# Apply the function to the Torque column
obj_df['Torque'] = obj_df['Torque'].apply(convert_torque)
```

```python
def categorize_brake_type(value):
    value = str(value).lower()  # Convert the value to lowercase for easier comparison
    
    # Disc brakes category
    if 'disc' in value or 'ventilated' in value:
        return 'Disc Brakes'
    
    # Drum brakes category
    elif 'drum' in value:
        return 'Drum Brakes'
    
    # Other brake types
    else:
        return 'Other Brake Types'
```

```python
obj_df['Rear Brake Type'] = obj_df['Rear Brake Type'].apply(categorize_brake_type)
obj_df['Front Brake Type'] = obj_df['Front Brake Type'].apply(categorize_brake_type)
```

```python
def convert_acceleration(value):
    # Convert the value to string to handle various formats
    value = str(value).lower()
    
    # Remove non-numeric characters ('sec', 's', etc.) and keep only numbers and decimal point
    value = ''.join(c for c in value if c.isdigit() or c == '.')
    
    # Convert the cleaned value to a float
    try:
        return float(value)
    except ValueError:
        return None  # Return None (or NaN) if conversion fails

# Apply the function to the 'Acceleration' column
obj_df['Acceleration'] = obj_df['Acceleration'].apply(convert_acceleration)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
x = df_all.drop("price",axis=1)
y = df_all["price"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model = RandomForestRegressor()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
R2score = r2_score(y_test,y_pred)
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
print(f"R2score :{R2score}")
print(f"meanabsolute_error :{MAE}")
print(f"Meansqua_error is:{MSE}")
```

```python
feature_importances =model.feature_importances_

# Create a DataFrame to store feature names and their importances
importance_df = pd.DataFrame({
    'Feature': x.columns,
    'Importance': feature_importances
})

# Sort features by importance
sorted_importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Rearrange columns of the original DataFrame based on importance
sorted_columns = sorted_importance_df['Feature'].tolist()
sorted_df = df_all[sorted_columns]

# Display top 10 important features
print(sorted_importance_df.head(20))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                           param_grid=param_grid,
                           scoring='r2',
                           cv=5,
                           verbose=2,
                           n_jobs=-1)
grid_search.fit(x_train, y_train)

best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```
